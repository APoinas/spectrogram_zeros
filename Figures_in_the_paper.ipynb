```python
import numpy as np

from Signal import Signal

def Remove_axis(ax):
    ax.xaxis.set_tick_params(labelbottom=False)
    ax.yaxis.set_tick_params(labelleft=False)
    ax.set_xticks([])
    ax.set_yticks([])
```

# Hermite functions

```python
np.random.seed(100)
sig = Signal({"SNR":1000000, "k":1})
```

```python
fig, ax = sig.plot(plot_type=["theo"])
fig.set_figwidth(5)
fig.set_figheight(5)
ax[0].set_xlim(-3, 3)
ax[0].set_ylim(-3, 3)
ax[0].plot(0, 0, 'o', color="white")
Remove_axis(ax[0])
```

```python
fig, ax = sig.plot(plot_type=["spectro"])
fig.set_figwidth(5)
fig.set_figheight(5)
ax[0].set_xlim(-3, 3)
ax[0].set_ylim(-3, 3)
Remove_axis(ax[0])
```

```python
np.random.seed(100)
sig = Signal({"SNR":400, "k":10})
```

```python
fig, ax = sig.plot(plot_type=["spectro"])
Remove_axis(ax[0])
```

```python
fig, ax = sig.plot(plot_type=["intensity"])
Remove_axis(ax[0])
```

# One chirp

```python
np.random.seed(200)
sig = Signal({"SNR":100, "a":-5, "b":0.4})
```

```python
fig, ax = sig.plot(plot_type=["spectro"])
ax[0].set_ylim(-4, 6)
Remove_axis(ax[0])
```

```python
fig, ax = sig.plot(plot_type=["intensity"])
ax[0].set_ylim(-4, 6)
Remove_axis(ax[0])
```

# Two parallel chirps

```python
np.random.seed(300)
sig = Signal({"SNR":[100, 40], "a":[-1, 0], "b":0.4}, T_length=15)
```

```python
fig, ax = sig.plot(show_max=True, plot_type=["theo"])
Remove_axis(ax[0])
sig.add_yellow_box(ax[0], -7)
```

```python
fig, ax = sig.plot(show_max=True, plot_type=["spectro"])
Remove_axis(ax[0])
sig.add_yellow_box(ax[0], -7)
```

```python
fig, ax = sig.plot(show_max=True, plot_type=["intensity"])
Remove_axis(ax[0])
```
